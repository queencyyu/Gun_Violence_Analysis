#### 2. Data Quality & Consistency Check

```python
# Checking mixed type data columns

for col in df.columns.tolist():
    mix = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis=1)
    if len(df[mix]) > 0 :
        print (col)
```

```python
# Changing data type for mixed type columns

df['address'] = df['address'].astype('str')
df['source_url'] = df['source_url'].astype('str')
df['gun_stolen'] = df['gun_stolen'].astype('str')
df['gun_type'] = df['gun_type'].astype('str')
df['incident_characteristics'] = df['incident_characteristics'].astype('str')
df['location_description'] = df['location_description'].astype('str')
df['notes'] = df['notes'].astype('str')
df['participant_age'] = df['participant_age'].astype('str')
df['participant_age_group'] = df['participant_age_group'].astype('str')
df['participant_gender'] = df['participant_gender'].astype('str')
df['participant_name'] = df['participant_name'].astype('str')
df['participant_relationship'] = df['participant_relationship'].astype('str')
df['participant_status'] = df['participant_status'].astype('str')
df['participant_type'] = df['participant_type'].astype('str')
df['sources'] = df['sources'].astype('str')
```

```python
# Checking missing data values

df.isnull().sum()
```

```python
# Imputting some of the missing string value as 'Unknown'

df['address'].fillna('Unknown', inplace = True)
df['source_url'].fillna('Unknown', inplace = True)
df['gun_stolen'].fillna('Unknown', inplace = True)
df['gun_type'].fillna('Unknown', inplace = True)
df['incident_characteristics'].fillna('Unknown', inplace = True)
df['location_description'].fillna('Unknown', inplace = True)
df['participant_age'].fillna('Unknown', inplace = True)
df['participant_age_group'].fillna('Unknown', inplace = True)
df['participant_gender'].fillna('Unknown', inplace = True)
df['participant_relationship'].fillna('Unknown', inplace = True)
df['participant_status'].fillna('Unknown', inplace = True)
df['participant_type'].fillna('Unknown', inplace = True)

# The of rest of the columns will be dropped due to no relation with the analysis, so the data will not be imputted.
```

```python
# Checking for duplicate data

df_duplicates = df[df.duplicated()]
df_duplicates
```

#### 3. Data Wrangling

```python
# Dropping unneeded columns for analysis

df = df.drop(columns=['incident_url', 'source_url', 'incident_url_fields_missing', 'congressional_district', 'gun_type', 'location_description',
                      'notes','participant_name', 'sources','state_house_district','state_senate_district','gun_stolen'])
df.head()
```

```python
# Convert 'date' column into datatime format

df['date'] = pd.to_datetime(df['date'])

# Exctract day of the week, day, month, and year

df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['day_of_week'] = df['date'].dt.day_name()
```

```python
# Set max display on column

pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
# Creating a function to categorise events based on the 'incident_characteristics'

def categorise_event(incident_characteristics):
    if 'Gang involvement' in incident_characteristics:
        return 'Gang involvement'
    elif 'Suicide' in incident_characteristics:
        return 'Suicide'
    elif 'Armed robbery' in incident_characteristics:
        return 'Armed robbery'
    elif 'Drug involvement' in incident_characteristics:
        return 'Drug involvement'
    elif 'Accidental shooting' in incident_characteristics:
        return 'Accidental shooting'
    elif 'Home Invasion' in incident_characteristics:
        return 'Home Invasion'
    elif 'Domestic Violence' in incident_characteristics:
        return 'Domestic Violence'
    elif 'Bar/club incident' in incident_characteristics:
        return 'Bar/club incident'
    elif 'Accidental Shooting' in incident_characteristics:
        return 'Accidental Shooting'
    elif 'Gun at school' in incident_characteristics:
        return 'Gun at school'
    elif 'Defensive Use' in incident_characteristics:
        return 'Defensive Use'
    elif 'Car-jacking' in incident_characteristics:
        return 'Car-jacking'
    else:
        return 'Others'

# Creating the 'incident_event' column based on the categorisation function

df['incident_event'] = df['incident_characteristics'].apply(categorise_event)

df.incident_event.value_counts()
```

```python
# Creating a function to categorise shots/no shots happeneddef categorise_event(incident_characteristics):

def categorise_shots(incident_characteristics):
    if 'Non-Shooting Incident' in incident_characteristics:
        return 'Non-Shooting Incident'
    elif 'Shot - Dead' in incident_characteristics:
        return 'Shot - Dead'
    elif 'Shot - Wounded/Injured' in incident_characteristics:
        return 'Shot - Wounded/Injured'
    elif 'Shots Fired - No Injuries' in incident_characteristics:
        return 'Shots Fired - No Injuries'
    else:
        return 'Unknown'
    
# Creating the 'incident_shot' column based on the categorisation function

df['incident_shot'] = df['incident_characteristics'].apply(categorise_shots)

df.incident_shot.value_counts()
```

```python
df.head()
```

```python
# extract data from participant_age_group column into new columns

def extract_age_group_column(age_group):
    df['n_adult'] = df[age_group].str.count('Adult')
    df['n_teen'] = df[age_group].str.count('Teen')
    df['n_child'] = df[age_group].str.count('Child')
    return df

df = extract_age_group_column('participant_age_group')

df.head()
```

```python
# extract data from participant_gender column into new columns

def extract_gender_column(gender):
    df['n_male'] = df[gender].str.count('Male')
    df['n_female'] = df[gender].str.count('Female')
    return df

df = extract_gender_column('participant_gender')

df.head()
```

```python
# extract data from participant_type column into new columns

def extract_type_column(type):
    df['n_victim'] = df[type].str.count('Victim')
    df['n_suspect'] = df[type].str.count('Suspect')
    return df

df = extract_type_column('participant_type')

df.head()
```

```python
# Dropping unused columns

df = df.drop(columns=['incident_characteristics', 'participant_age', 'participant_age_group', 'participant_gender',
                      'participant_relationship', 'participant_status', 'participant_type'])
             
df.head()
```

```python
# Reorganising columns order

df = df[['incident_id', 'date','day_of_week', 'day', 'month','year','state', 'city_or_county', 'address','latitude',
        'longitude', 'incident_event', 'incident_shot','n_guns_involved', 'n_killed', 'n_injured','n_adult',
        'n_teen', 'n_child', 'n_male', 'n_female', 'n_victim', 'n_suspect']]

df.head()
```
